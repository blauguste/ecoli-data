#### Use the distribution of # genomes per sRNA to determine that a large proportion of genomes have 73 or more

```python
sums = pae.append(pd.Series(pae.sum(), name='Total'))
sums
```

```python
sums['ct'] = pae.sum(axis=1)
sums
```

```python
asmbls = sums.loc[sums['ct'] >= 73].index.tolist()
```
